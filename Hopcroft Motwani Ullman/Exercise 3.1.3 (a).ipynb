```python
from automata.fa.nfa import NFA
from automata.fa.dfa import DFA
from helpers import get_automaton_display
```

```python
start_state = 'a'
states = {'a','b','c', 'e'}
state_transitions = {'a': {'0': {'a'}, '1': {'a', 'b'}},
                     'b': {'0': {'c'}, '1': {}},
                     'c': {'0': {}, '1': {'e'}},
                     'e': {'0': {'e'}, '1': {'e'}}}

nfa_complement = NFA(
    states=states,
    input_symbols={'0', '1'},
    transitions=state_transitions,
    initial_state=start_state,
    final_states={'e'})
```

```python
get_automaton_display(nfa_complement)
```

```python
dfa = DFA.from_nfa(nfa_complement)
dfa = dfa.complement()
get_automaton_display(dfa)
```

```python
nfa = NFA.from_regex('0*1*(100+1*)*(10)?', input_symbols={'0','1'})
```

```python
dfa = DFA.from_nfa(nfa)
get_automaton_display(dfa)
```
