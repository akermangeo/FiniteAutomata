Algorithm for processing regular expressions in linear time

```python
class string:
    pass

class star(string):
    def __init__(self, expression):
        self._expression = expression

    def Print(self):
        self._expression.Print()
        print("*")

class brackets(string):
    def __init__(self, expression):
        self._expression = expression
    
    def Print(self):
        print("(")
        self._expression.Print()
        print(")")

class union(string):
    def __init__(self, expression1, expression2):
        self._expression1 = expression1
        self._expression2 = expression2

    def Print(self):
        self._expression1.Print()
        print('|')
        self._expression2.Print()

class concat(string):
    def __init__(self, expression1, expression2):
        self._expression1 = expression1
        self._expression2 = expression2

    def Print(self):
        self._expression1.Print()
        print('.')
        self._expression2.Print()

class leaf(string):
    def __init__(self, letter):
        self.letter = letter

    def Print(self):
        print(self.letter)

class null_expression():
    def Print(self):
        pass

```

```python
class regex:
    def __init__(self, string):
        self.expression = string

    def next_character(self):
        next_character = self.expression[0]
        self.expression = self.expression[1:]
        return next_character
    
    def peek_for_close_bracket(self):
        if not self.empty() and self.expression[0] == ")":
            return True
        return False
    
    def empty(self):
        return len(self.expression) == 0
    
    def replace_bracket(self):
        self.expression = ")" + self.expression
```

```python
def parse(regex, prev_expression = null_expression()):
    if regex.empty() or regex.peek_for_close_bracket():
        return prev_expression
    
    character = regex.next_character()
    
    
    if character == 'a' or character == 'b':
        if isinstance(prev_expression, null_expression):
            leef_expression = leaf(character)
            return parse(regex, leef_expression)
        
        raise RuntimeError("Ill formed expression")
    
    if character == '|':
        return parse(regex, union(prev_expression, parse(regex)))
    
    if character == '.':
        return parse(regex, concat(prev_expression, parse(regex)))
    
    if character == '*':
        star_expression = star(prev_expression)
        return parse(regex, star_expression)
    
    if character == "(":
        if isinstance(prev_expression, null_expression):
            bracket_expression = brackets(parse_bracket(regex, null_expression()))
            return parse(regex, bracket_expression)
        
        raise RuntimeError("Ill formed expression")
    

```

```python
def parse_bracket(regex, prev_expression = null_expression()):
    if regex.empty():
        raise RuntimeError("Not enough closing brackets")
    
    character = regex.next_character()

    if character == ")":
        return prev_expression
    
    if character == 'a' or character == 'b':
        if isinstance(prev_expression, null_expression):
            return parse_bracket(regex, leaf(character))
        raise RuntimeError("Ill formed expression")
    if character == '|':
        return parse_bracket(regex, union(prev_expression, parse(regex)))
    
    if character == '.':
        return parse_bracket(regex, concat(prev_expression, parse(regex)))
    
    if character == '*':
        return parse_bracket(regex, star(prev_expression))
    
    if character == "(":
        if isinstance(prev_expression, null_expression):
            bracket_expression = brackets(parse_bracket(regex, null_expression()))
            return parse_bracket(regex, bracket_expression)
        
        raise RuntimeError("Ill formed expression")
    

```
