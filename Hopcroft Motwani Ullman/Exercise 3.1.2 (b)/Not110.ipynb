```python
from automata.fa.nfa import NFA
from automata.fa.dfa import DFA
from IPython.core.display import SVG
```

```python
def print(automaton):
    diagram = automaton.show_diagram()
    img = diagram.create_svg()
    return SVG( data=img )
```

```python
start_state = 'a'
states = {'a','b','c','e'}
state_transitions = {'a': {'0': {'a'}, '1': {'a', 'b'}},
                     'b': {'0': {}, '1': {'c'}},
                     'c': {'0': {'e'}, '1': {}},
                     'e': {'0': {'e'}, '1': {'e'}}}

nfa_complement = NFA(
    states=states,
    input_symbols={'0', '1'},
    transitions=state_transitions,
    initial_state=start_state,
    final_states={'e'})
```

```python
print(nfa_complement)
```

```python
dfa = DFA.from_nfa(nfa_complement)
dfa = dfa.complement()
print(dfa)
```

```python
expected_nfa = NFA.from_regex('0*(10+)*1*', input_symbols={'0','1'})
```

```python
print(DFA.from_nfa(expected_nfa))
```
